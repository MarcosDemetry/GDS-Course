```python
import geopandas, osmnx
import matplotlib.pyplot as plt
import contextily as cx
```

```python
liv = geopandas.read_file("Census_Residential_Data_Pack_2011/Local_Authority_Districts/E08000012/shapefiles/E08000012.shp")
```

```python
bikepaths = osmnx.graph_from_place("Liverpool, UK", network_type="bike")
```

```python
pubs = osmnx.geometries.geometries_from_place("Liverpool, UK",
                                  tags={"amenity": "pub"}
                                 )
```

# Task 1

With those three layers, try to complete the following tasks:

Make a map of the Liverpool neighborhoods that includes the following characteristics:
* Features a title
* Does not include axes frame
* It has a figure size of 10 by 11
* Polygons are all in color "#525252" and 50% transparent
* Boundary lines (“edges”) have a width of 0.3 and are of color "#B9EBE3"
* Includes a basemap with the Stamen watercolor theme

```python
providers = {}

def get_providers(provider):
    if "url" in provider:
        providers[provider['name']] = provider
    else:
        for prov in provider.values():
            get_providers(prov)

get_providers(cx.providers)
```

```python
liv_wm = liv.to_crs(epsg=3857)
```

```python
# Setup figure and axis
f, ax = plt.subplots(1, figsize=(10, 11))

# Add layer of polygons on the axis
liv_wm.plot(ax=ax,
         alpha = 0.5,
         color='#525252',
         edgecolor='#B9EBE3',
         linewidth=0.3)

osmnx.graph_to_gdfs(bikepaths, nodes=False).to_crs(liv_wm.crs)\
     .plot(ax=ax, 
           linewidth=0.3, 
           color="#B9EBE3")

pubs.to_crs(liv_wm.crs).plot(ax=ax, color="yellow", markersize=1)

# Add figure title
f.suptitle("Liverpool: Pubs and Bike Paths")
# Set axis off
ax.set_axis_off()
cx.add_basemap(ax, source=providers['Stamen.Watercolor']);

plt.savefig('Liverpool_pubs_and_bike_paths.png')
```

# Task 3: The gender gap on the streets

Which group accounts for longer total street length in Zaragoza: men or women? By how much?

```python
url = ("https://github.com/geochicasosm/lascallesdelasmujeres"\
       "/raw/master/data/zaragoza/final_tile.geojson")
url
```

```python
streets = geopandas.read_file(url)
```

First, let's inspect the data

```python
streets.info()
```

```python
streets.plot()
```

We see that we have a gender indicator. So, I'm thinking that we do the following:
- Calculate street lengths per person
- Calculate total street lengths per gender
- Calculate difference in total streeet lengths between men and women

```python
# Calculating street length
street_length = streets.to_crs(epsg=25830).length
```

```python
street_length_gender =  streets.to_crs(epsg=25830).length
streets['length'] = street_length_gender
```

```python
streets_lengths_gender = streets[['gender','length']].groupby('gender').sum()
```

```python
streets_lengths_gender.head()
```

```python
male = streets_lengths_gender[streets_lengths_gender.index=="Male"].values[0][0]/ 1000
female = streets_lengths_gender[streets_lengths_gender.index=="Female"].values[0][0]/ 1000
```

```python
print(f"Males occupy a street length of{male: .2f} KM and females occupy a street length of{female: .2f} KM.\nThis is a difference of{male-female: .2f} KM.")
```
